# While and For Loop

```python
i = 0
while i != 5:
    print('i is: ',i)
    i += 1
print(i,' is equal to 5')
```

```python
lis = [1,2,3,4,5]
for i in lis:
    print('i is: ',i)
print('')

for index,value in enumerate(lis):
    print(index, " : ",value)
print('')

dictionary = {'spain':'madrid','france':'paris'}
for key,value in dictionary.items():
    print(key," : ",value)
print('')

for index,value in data[['Attack']][0:1].iterrows():
    print(index," : ",value)
```

# Nested Function

```python
def square():
    def add():
        x = 2
        y = 3 
        z = x + y
        return z
    
    return add()**2
print(square())
```

```python
def f(a,b=1,c=2):
    y = a+b+c
    return y
print(f(5))
print(f(5,4,3))
```

```python
#flexible arguments *args

def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)
```

```python
def f(**kwargs):
    
    for key,value in kwargs.items():
        print(key," ",value)
    f(country='spain',captital = 'madrid',population=123456)
```

# List Comprehension

```python
num1 = [1,2,3]
num2 = [i + 1 for i in num1]
print(num2)
```

```python
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i<7 else i+5 for i in num1]
print(num2)
```
